```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
#dataset --> https://www.kaggle.com/datasets/yasserh/housing-prices-dataset
df = pd.read_csv("../DataSet/Housing.csv")
```

```python
data = pd.get_dummies(df, drop_first=True)
```

```python
data.head()
```

```python
x = data.iloc[ : , 1: ]
```

```python
x.head()
```

```python
y = data.iloc[: , 0]
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0) # herea random state help to give same result of split when ever run the code
```

# Train a Model

```python
model = SVR(kernel="linear")
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.support_
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```
